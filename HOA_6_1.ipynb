# **Without using libraries**

**MEAN**

```python
money = salaries

len(salaries) # Getting the lenght of the dataframe salaries
```

```python
means = sum(salaries) / len(salaries) #function of mean without using a library
print(means)
```

 **MEDIAN**

```python
money.sort()
x = money[49] + money[50] # There are two middle values in the data so we get the mean
medianz = x / 2 #

print(medianz)
```

**MODE**

```python
money.sort()
check=[] # Creating a list to store frequency

# Iterating through the list
i=0
while i<len(money):
  check.append(money.count(money[i]))  # Count the occurrences of the current salary in the list and append to list
  i+=1

dic = dict(zip(money, check)) # Create a dictionary to map salaries to their frequencies
dic2 = {k for (k,v)in dic.items() if v== max(check)} # Create a set to store salaries with the maximum frequency

print(dic2)
```

**SAMPLE VARIANCE**

```python
summation = 0
for x in salaries:
  summation += (x - mean)**2

samp_var = summation/(len(salaries)-1)

print(samp_var)
```

**STANDARD DEVIATION**

```python
std_dvn = samp_var**0.5
print(std_dvn)
```

# **With using libraries**

```python
from statistics import * # Importing the whole statistic library
sa = salaries
```

```python
# Calculating and comparing the mean with and without library
mean_lib = mean(sa)

print(mean_lib, means)

if means == mean_lib:
  print('same')
```

```python
# Calculating and comparing the median
mdn = median(sa)

print(mdn, medianz)

if mdn == medianz:
  print('same')
```

```python
# Calculation and comparing the mode
md = mode(sa)

print(md, dic2) # They are the same
```

```python
# Calculating and comparing sample variance
vari = variance(salaries)

print(vari, samp_var)

if vari == samp_var:
  print('same')
```

```python
# Calculating and comparing the standard deviation
stddv = stdev(salaries)
print(stddv, std_dvn)

if stddv == std_dvn:
  print('same')
```

**Exercise 2**

Using the same data, calculate the following statistics using the functions in the statistics module where appropriate:

- Range
- Coefficient of variation Interquartile range
- Quartile coefficient of dispersion


```python
# Calculating the range of the dataset
max(money) - min(money)
```

```python
# Calculating the coefficient of variation interquartile range
def calculate_cv_iqr(sal):

  # Calculate quartiles
  q1 = median(sal[:len(sal) // 2])
  q3 = median(sal[len(sal) // 2:])

  # Handle equal quartile case for IQR
  iqr = None
  if q1 != q3:
    iqr = q3 - q1

  return iqr

# Calculate IQR
iqr = calculate_cv_iqr(money)

# Print the results
print("Interquartile Range (IQR):", iqr)
```

```python
def calculate_qcd(money):

  # Calculate quartiles
  q1 = median(money[:len(money) // 2])
  q3 = median(money[len(money) // 2:])

  # Calculate QCD
  qcd = (q3 - q1) / (2 * mdn)

  return qcd

# Calculate QCD
qcd = calculate_qcd(money)

# Print the result
print("Quartile Coefficient of Dispersion (QCD):", qcd)
```
