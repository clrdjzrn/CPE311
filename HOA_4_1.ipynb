```python
def TOH(n, Src, Dstn, Aux):
    if n == 1: #Base case: if there is only one disk print the move and return
        print(f"Move disk 1 from {Src} to {Dstn}")
        return

    TOH(n-1, Src, Aux, Dstn)  #Recursive case: if there are more than one disk, move the top n-1 disks from the source to the aux
    print(f"Move disk {n} from {Src} to {Dstn}")
    TOH(n-1, Aux, Dstn, Src)

n = 5
TOH(n, 'A', 'C', 'B')


```

Observation:

The function TOH is called recursively to move the disks between the rods. The base case stops the recursion when there is only one disk left. The recursive call ensures that smaller disks are moved to the auxiliary rod before moving the largest.
